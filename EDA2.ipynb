# ðŸ“Œ 2. **Univariate Analysis**

Univariate analysis helps us understand the distribution of each variable individually.

âœ… For **numerical columns**, we used:
- Histograms (with KDE) to visualize the distribution.
- Boxplots to identify outliers and spread.

âœ… For **categorical columns**, we used:
- Count plots for frequency distribution.
- For long-label columns (e.g., `Title`), we used **horizontal bar charts** to improve readability.

ðŸ“Œ **Shapiro-Wilk Test** was applied to test for normality:
- Most numerical columns are **not normally distributed**, as expected in real-world datasets.

```python
# Univariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

sns.set(style="whitegrid")

numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print("\nShapiro-Wilk Test Results (first 500 rows per column):")
for col in numerical_cols:
    stat, p = shapiro(df[col].head(500))
    print(f"{col}: p-value = {p:.4e}")

def label_numeric_axis(col_name, data_series):
    if pd.api.types.is_numeric_dtype(data_series):
        if data_series.max() <= 1 and data_series.min() >= 0:
            return f"{col_name} (Scaled 0â€“1)"
        else:
            return col_name
    else:
        return col_name

# Histograms & Boxplots
for col in numerical_cols[:3]:
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    sns.histplot(df[col], bins=30, kde=True, ax=ax[0])
    ax[0].set_title(f'Histogram of {col}')
    ax[0].set_xlabel(label_numeric_axis(col, df[col]))
    ax[0].set_ylabel("Count")
    
    sns.boxplot(x=df[col], ax=ax[1])
    ax[1].set_title(f'Boxplot of {col}')
    ax[1].set_xlabel(label_numeric_axis(col, df[col]))
    
    plt.tight_layout()
    plt.show()
```

The histogram illustrating the distribution of video duration reveals several key insights. The majority of videos have a very short duration, indicating a strong preference for short-form content among the videos in the dataset.This suggests that while most videos are short, there are a small number of longer videos that exist.. Overall, the histogram highlights a clear preference for shorter video durations, suggesting that content designed to be brief may attract more viewers in this dataset.

# ðŸ“Œ 3. **Categorical Analysis**

We analyzed categorical columns to understand the frequency of categories.

âœ… For columns with **many unique values** (e.g., Titles), we plotted only the **Top 10** categories.

âœ… We used **horizontal bar charts** for better readability of long labels.

```python
df['Region'].value_counts().head(11)
```

```python
# ðŸ“Œ Categorical Analysis (Improved Visualization â€“ Excluding 'Video ID')
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove 'Video ID' from the list
if 'Video ID' in categorical_cols:
    categorical_cols.remove('Video ID')

max_categories = 10

for col in categorical_cols:
    n_unique = df[col].nunique()
    print(f"\nCategorical Column: {col}, Unique Values: {n_unique}")
    
    if n_unique > max_categories:
        top_vals = df[col].value_counts().nlargest(max_categories).index
        data_filtered = df[df[col].isin(top_vals)]
        
        plt.figure(figsize=(10, 6))
        sns.countplot(data=data_filtered, y=col, order=top_vals)
        plt.title(f'Count Plot of {col} (Top {max_categories} of {n_unique})')
        plt.xlabel("Count")
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
    else:
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, y=col, order=df[col].value_counts().index)
        plt.title(f'Count Plot of {col} (All {n_unique})')
        plt.xlabel("Count")
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
```

### ðŸ“Œ **Time-based Analysis â€“ Hourly Trends**

This analysis helps explore **when videos are published throughout the day**, and whether specific hours correlate with higher user engagement.

- We extract the hour from the `'Publication Time'` column.
- Then, we calculate total **views, likes, and comments** per hour.
- Bar plots show the distribution of interactions across the 24 hours of a day.

**Interpretation Tip:**  
High interaction during certain hours (e.g., afternoon or evening) may indicate better engagement timing and can inform video posting strategies.


```python
# ðŸ“Œ Extract Hour from Publication Time
df['Hour'] = df['Publication Time'].dt.hour

# Group by Hour and compute total views, likes, comments
stats_by_hour = df.groupby('Hour').agg({
    'Views': 'sum',
    'Likes': 'sum',
    'Comments': 'sum'
}).reset_index()

# Plot: Views by Hour
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Views', data=stats_by_hour)
plt.title('Total Views by Hour of the Day')
plt.xlabel('Hour (0 = Midnight)')
plt.ylabel('Total Views')
plt.tight_layout()
plt.show()

# Plot: Likes by Hour
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Likes', data=stats_by_hour)
plt.title('Total Likes by Hour of the Day')
plt.xlabel('Hour (0 = Midnight)')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()

# Plot: Comments by Hour
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Comments', data=stats_by_hour)
plt.title('Total Comments by Hour of the Day')
plt.xlabel('Hour (0 = Midnight)')
plt.ylabel('Total Comments')
plt.tight_layout()
plt.show()

```

# ðŸ“Œ 5. **Multivariate Analysis (Parallel Coordinates Plot)**

Parallel coordinates plots are used to explore multivariate relationships across several numerical features simultaneously.

âœ… We scaled the numeric columns using **MinMaxScaler** to bring them into a [0â€“1] range for better comparison.

âœ… Instead of using `Video ID` (which has too many unique values), we used **`Region`** as the class label for grouping the lines in the plot.

ðŸ“Œ This makes the plot more readable and highlights patterns across groups rather than individual videos.

```python
# Multivariate Analysis (Parallel Coordinates Plot)
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import MinMaxScaler

class_col = "Region"  # âœ… Use a good class column, not Video ID

if class_col in df.columns and class_col in categorical_cols:
    sample_size = 500
    sample_df = df.sample(n=sample_size, random_state=42) if df.shape[0] > sample_size else df.copy()

    parallel_cols = numerical_cols[:5]
    scaler = MinMaxScaler()
    sample_df_scaled = sample_df.copy()
    sample_df_scaled[parallel_cols] = scaler.fit_transform(sample_df_scaled[parallel_cols])
    sample_df_scaled[class_col] = sample_df_scaled[class_col].astype(str)

    plt.figure(figsize=(12, 6))
    parallel_coordinates(sample_df_scaled[[class_col] + parallel_cols], class_col, colormap='Set2')
    plt.title(f"Parallel Coordinates Plot (Grouped by {class_col})")
    plt.tight_layout()
    plt.show()
else:
    print(f"âš  Column '{class_col}' not found or not suitable for grouping in parallel coordinates.")
```

### **How does the use of a particular language affect interaction?**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# extract column of language
language_cols = [col for col in df.columns if "Language_" in col]


language_views = df[language_cols].multiply(df["Views"], axis=0).sum()
language_comments = df[language_cols].multiply(df["Comments"], axis=0).sum()
language_likes = df[language_cols].multiply(df["Likes"], axis=0).sum()


language_stats = pd.DataFrame({
    "Views": language_views,
    "Comments": language_comments,
    "Likes": language_likes
})


plt.figure(figsize=(12, 6))
language_stats.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightpink', 'lightgreen'])

plt.title("Comparison of Views, Comments, and Likes per Language")
plt.xlabel("Language")
plt.ylabel("Total Count of views, comment and like")
plt.xticks(rotation=90)
plt.legend(["Views", "Comments", "Likes"])
plt.show()
```

From the graph, we can see that the use of a specific language can significantly increase interaction by raising the number of views, likes, and comments. It is evident that the English language accounts for the highest levels of interaction, featuring a substantial number of views, likes, and comments. Following English, the Spanish language also shows considerable engagement, though it is lower than that of English.

This indicates that the choice of language plays a crucial role in the reach and interaction of content. Additionally, the graph reveals that languages like Russian and Hindi have much lower interaction levels. This may reflect various factors such as audience size and cultural relevance, which can affect how content is received and engaged with by different demographic groups.

In summary, the graph illustrates that content in widely spoken languages, particularly English and Spanish, tends to attract more interaction, while content in languages with smaller audiences, such as Russian and Hindi, may face challenges in achieving similar engagement levels.
